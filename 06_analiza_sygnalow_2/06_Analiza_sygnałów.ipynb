```python
import numpy as np
import matplotlib.pyplot as plt
import aseegg as ag
```

###   Zadanie 1 <img src="../images/pencil.png" width="20" align="left">

Narysuj *nowy_sygnal* w dziedzinie czasu i częstotliwości z użyciem subplotów. 

```python
f1 = 5
f2 = 17
f3 = 31
czestProbkowania = 250
t = 3


t_axis = np.linspace(0, t, t * czestProbkowania)
nowy_sygnal = np.sin(2* np.pi* f1 *t_axis)+np.sin(2* np.pi*f2 *t_axis)+np.sin(2* np.pi* f3 *t_axis)
```

## Filtracja

```python
przefiltrowany = ag.gornoprzepustowy(nowy_sygnal, czestProbkowania, 10)
```

## Jak wygląda wcześniej stworzony sygnał "*nowy_sygnal*"  po filtracji? 

```python
f_axis = np.linspace(0, czestProbkowania, t * czestProbkowania)
przefiltrowany_ffted = ag.FFT(przefiltrowany)


plt.subplot(2,1,1)
plt.plot(t_axis, przefiltrowany)

plt.subplot(2,1,2)
plt.plot(f_axis, przefiltrowany_ffted)
plt.xlim([0,40])
```
