# Pobranie danych

### Instrukcja zbierania danych z OpenBCI Ganglion

1. Przenieś swój program do folderu 09_zbieranie_danych.
2. Zainstaluj pakiety:
>pip install brainflow
3. Podłącz dongle do USB (ten z zestawu Ganglion).
4. Sprawdź port COM:
    - Otwórz Menedżer urządzeń → Porty (COM i LPT)
    - Odczytaj numer (np. COM7, COM9).
    - Ustaw port COM w kodzie (edytuj plik data_collector.py):
    - COM_PORT = 'COM9'   # wpisz swój numer
5. Załóż elektrody:
    - CH1 → czoło
    - REF → ucho
    - D_G → drugie ucho
6. Podłącz baterie i włącz płytkę Ganglion.
7. Uruchom program w konsoli:
>python data_collector.py
8. Uruchom swój program (wyświetlacz liter lub obrazków) w osobym terminalu.
9. Zakończ zbieranie danych:
    - Ctrl + C
    -(plik CSV zapisze się automatycznie)
10. Zakończ działanie swojego programu. 
11. Zachowaj zebrane dane.
