# Pobranie danych na sucho

### Instrukcja zbierania danych z OpenBCI Ganglion

1. Zainstaluj pakiety:
>pip install brainflow
2. Podłącz dongle do USB (ten z zestawu Ganglion).
3. Sprawdź port COM:
    - Otwórz Menedżer urządzeń → Porty (COM i LPT)
    - Odczytaj numer (np. COM7, COM9).
    - Ustaw port COM w kodzie (edytuj plik data_collector.py):
    - COM_PORT = 'COM9'   # wpisz swój numer
4. Załóż elektrody:
    - CH1 → czoło
    - REF → ucho
    - D_G → drugie ucho
5. Podłącz baterie i włącz płytkę Ganglion.
6. Uruchom program w konsoli:
>python data_collector.py
7. Zakończ nagrywanie:
    - Ctrl + C
    -(plik CSV zapisze się automatycznie)
